```python
import pandas as pd
from datetime import datetime

def menu():
    while True:
        print("\n--- Menú Principal ---")
        print("1. Ver libros disponibles ")
        print("2. Registrar usuario ")
        print("3. Registrar libro ")
        print("4. Prestar libro ")
        print("5. Devolver libro ")
        print("6. Ver lista de libros prestados ")
        print("7. Ver lista de usuarios registrados ")
        print("8. Ver estadísticas de préstamo  ")
        print("9. Salir")

        opcion = input("Por favor, elige una opción: ")

        if opcion == "1":
            mostrar_libros()

        elif opcion == "2":
            nombre = input("Ingresa el nombre del usuario: ")
            cedula = int(input("Ingresa la Cédula del usuario: "))
            registrar_usuario(nombre, cedula)

        elif opcion == "3":
            titulo = input("Ingresa el título del libro: ")
            autor = input("Ingresa el autor del autor: ")
            registrar_libro(titulo, autor)

        elif opcion == "4":
            cedula = int(input("Ingresa la Cédula del usuario: "))
            titulo = input("Ingresa el título del libro: ")
            prestar_libro(cedula, titulo)

        elif opcion == "5":
            cedula = int(input("Ingresa la Cédula del usuario: "))
            titulo = input("Ingresa el título del libro: ")
            devolver_libro(cedula, titulo)

        elif opcion == "6":
            mostrar_usuarios_prestados()

        elif opcion == "7":
            mostrar_usuarios()

        elif opcion == "8":
            mostrar_estadisticas()

        elif opcion == "9":
            print("Saliste del Sistema")
            break

        else:
            print("Opción no válida. Por favor, elige una opción correcta.")


libros = [
    {"titulo": "Harry Potter", "autor": "J.K. Rowling", "disponible": True},
    {"titulo": "El principito", "autor": "Antoine de Saint-Exupéry", "disponible": True},
    {"titulo": "La máquina del tiempo", "autor": "Herbert George Wells", "disponible": True},
    {"titulo": "El problema de los tres cuerpos", "autor": "Liu Cixin", "disponible": True},
    {"titulo": "El Señor de los Anillos", "autor": "J.R.R. Tolkien", "disponible": True},
    {"titulo": "Fundacion", "autor": "Isaac Asimov", "disponible": True},
    {"titulo": "Una Cancion de Hielo y Fuego", "autor": "George R. R. Martin", "disponible": True},
    {"titulo": "Yo Robot", "autor": "Isaac Asimov", "disponible": True},
    {"titulo": "Crónica del Asesino de Reyes", "autor": "Patrick Rothfuss", "disponible": True},
    {"titulo": "Soy Leyenda", "autor": "JRichard Matheson", "disponible": True},
]

usuarios = [
    {"nombre": "David Gómez", "cedula": 12345, "prestamos": []},
    {"nombre": "Camilo Cifuentez ", "cedula": 67890, "prestamos": []},
    {"nombre": "Luis Paz", "cedula": 98765, "prestamos": []},
    {"nombre": "Diana Sarralde", "cedula": 54321, "prestamos": []},
]

prestamos = []
conteo_prestamos_libros = {}
conteo_prestamos_usuarios = {}

def mostrar_libros():
    print("\nLista de Libros Disponibles:")
    for libro in libros:
        if libro['disponible']:
            print(f"- {libro['titulo']} de {libro['autor']}")

def registrar_libro(titulo, autor):
    libro = {"titulo": titulo, "autor": autor, "disponible": True}
    libros.append(libro)
    print(f"El libro '{titulo}' ha sido registrado.")

def registrar_usuario(nombre, cedula):
    usuario_existente = None
    for u in usuarios:
        if u["cedula"] == cedula:
            usuario_existente = u
            break
    if usuario_existente:
        print(f"El usuario con Cédula {cedula} ya está registrado.")
    else:
        usuario = {"nombre": nombre, "cedula": cedula, "prestamos": []}
        usuarios.append(usuario)
        print(f"El usuario {nombre} con Cédula {cedula} ha sido registrado.")

def prestar_libro(cedula, titulo):
    usuario = None
    for u in usuarios:
        if u["cedula"] == cedula:
            usuario = u
            break

    libro = None
    for l in libros:
        if l["titulo"].lower().strip() == titulo.lower().strip() and l["disponible"]:
            libro = l
            break

    if usuario and libro:
        libro["disponible"] = False
        prestamo = {"usuario_cedula": cedula, "libro_titulo": titulo, "fecha_prestamo": datetime.now(), "fecha_devolucion": None}
        usuario["prestamos"].append(prestamo)
        prestamos.append(prestamo)


        conteo_prestamos_libros[titulo] = conteo_prestamos_libros.get(titulo, 0) + 1
        conteo_prestamos_usuarios[cedula] = conteo_prestamos_usuarios.get(cedula, 0) + 1

        print(f"El libro '{titulo}' ha sido prestado a {usuario['nombre']}. Fecha de préstamo: {prestamo['fecha_prestamo'].strftime('%d/%m/%Y %H:%M:%S')}.")
    else:
        print("No se puede realizar el préstamo. Verifica que el libro esté disponible y el usuario esté registrado.")


def devolver_libro(cedula, titulo):
    usuario = None
    for u in usuarios:
        if u["cedula"] == cedula:
            usuario = u
            break

    if usuario:
        prestamo_encontrado = None
        for p in usuario["prestamos"]:
            if p["libro_titulo"] == titulo and p["fecha_devolucion"] is None:
                prestamo_encontrado = p
                break

        if prestamo_encontrado:
            prestamo_encontrado["fecha_devolucion"] = datetime.now()
            libro = None
            for l in libros:
                if l["titulo"] == titulo:
                    libro = l
                    break
            if libro:
                libro["disponible"] = True
            print(f"El libro '{titulo}' ha sido devuelto por {usuario['nombre']}. Fecha de devolución: {prestamo_encontrado['fecha_devolucion'].strftime('%d/%m/%Y %H:%M:%S')}."
            )
        else:
            print("No se puede devolver el libro. Verifica que el libro esté prestado al usuario.")
    else:
         print("Usuario no encontrado.")


def mostrar_usuarios_prestados():
    for usuario in usuarios:
        if usuario["prestamos"]:
            libros_prestados = []
            for prestamo in usuario["prestamos"]:
              titulo_libro = prestamo["libro_titulo"]
              libros_prestados.append(titulo_libro)
            print(f"{usuario['nombre']} ha solicitado los siguientes libros: {', '.join(libros_prestados)}.")
        else:
            print(f"{usuario['nombre']} no ha solicitado libros.")

def mostrar_usuarios():
    print("\nLista de Usuarios Registrados:")
    for usuario in usuarios:
        print(f"- {usuario['nombre']}, Cédula: {usuario['cedula']}")

def mostrar_estadisticas():
    print("\n--- Estadísticas de Préstamo ---")

    if not conteo_prestamos_libros:
        print("No se han prestado libros todavía.")
    else:
        print("\nEstadísticas de Libros Prestados:")
        df_libros = pd.DataFrame(list(conteo_prestamos_libros.items()), columns=['Libro', 'Número de Préstamos'])
        df_libros_ordenado = df_libros.sort_values(by='Número de Préstamos', ascending=False)
        display(df_libros_ordenado)

    if not conteo_prestamos_usuarios:
        print("Ningún usuario ha prestado libros todavía.")
    else:
        print("\nEstadísticas de Usuarios con Más Préstamos:")
        df_usuarios = pd.DataFrame(list(conteo_prestamos_usuarios.items()), columns=['Cédula', 'Número de Préstamos'])

        user_names = {user['cedula']: user['nombre'] for user in usuarios}
        df_usuarios['Nombre de Usuario'] = df_usuarios['Cédula'].map(user_names)
        df_usuarios_ordenado = df_usuarios.sort_values(by='Número de Préstamos', ascending=False).drop('Cédula', axis=1)
        display(df_usuarios_ordenado)


menu()
```
