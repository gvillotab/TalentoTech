```python
import pandas as pd

value_n_list = ["Andrew", "Phillip", "John", "David", "Carlos", "Fabian", "Esteban", "Oscar", "Ricardo", "Luis"]
value_w_list = [50,        78,        91.5,   62,      71,       59,       80,        85,      66,        72] #[kg]
value_h_list = [1.76,      1.87,      1.69,   1.65,    1.70,     1.62,     1.72,      1.80,    1.68,      1.75] #[m]

data_dict = {"Name":value_n_list, "Weigh":value_w_list, "Height":value_h_list}

df = pd.DataFrame(data_dict)

value_b_list = []
cond_list = []
for i in range(10):
  bmi = value_w_list[i] / (value_h_list[i]**2)
  value_b_list.append(bmi)
  if bmi > 25:
    condicion = "Sobrepeso"  
  elif bmi < 18.5:
    condicion = "Bajo Peso" 
  else:
    condicion = "Saludable"
  cond_list.append(condicion)
df['BMI'] = value_b_list
df['Condicion'] = cond_list
df
```
